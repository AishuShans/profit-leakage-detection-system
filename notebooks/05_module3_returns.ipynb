returns and refund leakage analysis

import lib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

load cleaned dataset

```python
df = pd.read_csv("profit_leakage_cleaned.csv")
df.head()
```

returns overview

```python
df["return_flag"].value_counts()
```

return rate percentage

```python
return_rate = df["return_flag"].mean() * 100
return_rate
```

refund amount distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(df["refund_amount"], bins=40)
plt.title("Refund Amount Distribution")
plt.show()
```

refund vs non-returned profit

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x="return_flag",
    y="profit",
    data=df
)
plt.xticks([0,1], ["Not Returned", "Returned"])
plt.title("Profit Comparison: Returned vs Non-Returned Orders")
plt.show()

```

category-wise return rate

```python
category_return = df.groupby("product_category")["return_flag"].mean() * 100

plt.figure(figsize=(6,4))
category_return.plot(kind="bar")
plt.title("Return Rate by Product Category")
plt.ylabel("Return Rate (%)")
plt.show()
```

region-wise refund amount

```python
plt.figure(figsize=(6,4))
df.groupby("region")["refund_amount"].sum().plot(kind="bar")
plt.title("Total Refund Amount by Region")
plt.ylabel("Refund Amount")
plt.show()
```

revenue vs refund impact

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x="revenue",
    y="refund_amount",
    data=df,
    alpha=0.3
)
plt.title("Revenue vs Refund Amount")
plt.show()
```
